```python
import pandas as pd
import matplotlib.pyplot as plt #Use for visualization
import seaborn as sns #visualize regression nicelly
```

```python
df = pd.read_csv('temperatures.csv')
```

```python
df
```

```python
df.head()
```

```python
#input data
x = df['YEAR']

#output data
y = df['ANNUAL']
```

```python
#plt.figure(figsize=(16,9))
plt.title('Temperature Plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual Average Temperature')
plt.scatter(x,y)
```

```python
x = x.values
```

```python
x
```

```python
x = x.reshape(117, 1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x, y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
 regressor.predict([[2025]])
```

```python
#Assess the performance of regression models using MSE, MAE and R-Square metrics
predicted = regressor.predict(x)
```

```python
predicted
```

```python
y
```

```python
# Mean Absolute Error
import numpy as np
np.mean(abs (y-predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, predicted)
```

```python
# Mean Squared Error
np.mean((y-predicted) ** 2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, predicted)
```

```python
# R-Square Error
from sklearn.metrics import r2_score
r2_score(y, predicted)
```

```python
regressor.score(x,y)
```

```python
# Visualize the regression model
plt.title('Temperature Plot of INDIA')
plt.xlabel('Year')
plt.ylabel('Annual Average Temperature')
plt.scatter(x,y, label = 'actual', color='y')
plt.plot(x, predicted, label = 'predicted', color='b')
plt.legend()
```
