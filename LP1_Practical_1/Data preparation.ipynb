```python
import numpy as np
```

```python
# predicted 100 report as COVID positive, but only 45 of those were actually positive.
# Total 50 people were actually COVID positive.total 555 samples.
actual = list(np.ones(45))+list(np.zeros(55))+list(np.ones(5))+list(np.zeros(450))
np.array(actual)
```

```python
predicted = list(np.ones(100)) + list(np.zeros(455))
np.array(predicted)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_predictions(actual, predicted)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(actual, predicted))
```
